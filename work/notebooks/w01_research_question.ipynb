# Week 1 Research Question

## 1. My Lane (and Why)

**Selected Lane:** Predict pages that are likely to have low CTR despite receiving search impressions.

I selected this lane because improving click-through rate (CTR) can increase website traffic without creating additional content. The starter dataset contains search metrics such as impressions, clicks, CTR, and average position, making it suitable for analyzing which pages may need optimization.

## 2. The Question

**Research Question**

Can we identify webpages that are likely to have a low click-through rate (CTR) so they can be prioritized for optimization?

**Unit of Analysis**

One webpage.

**Model Output**

A prediction indicating whether a page is likely to have a low CTR.

**Decision**

Identify which pages should be reviewed first.

**Action**

Optimize titles, meta descriptions, or page content for the predicted low-CTR pages.

**Cost of a Wrong Recommendation**

If a high-performing page is incorrectly flagged, time may be wasted reviewing it. If a low-performing page is missed, an opportunity to improve traffic could be lost.

**Why Data or ML Can Help**

The dataset contains many pages with multiple search-related features. Machine learning can help identify useful patterns for prioritizing pages that deserve further review.
